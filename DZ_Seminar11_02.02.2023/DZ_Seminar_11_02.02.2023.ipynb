```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
f = -12 * x**4 * sin(cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
f
```

```python
plot(f, (x,-100000, 100000))
```

```python
solveset(f, x)
```

```python
nonlinsolve([f], x)
```
